# Parts 9–10 Demo: Batch Processing and Final Submission Assets

This notebook shows the final project layer: sector-like batch execution, resume-safe outputs, final candidate catalog, visual summaries, and report assets.


```python
from pathlib import Path
import sys
ROOT = Path('..').resolve()
sys.path.insert(0, str(ROOT / 'src'))

from exoplanet_pipeline.config import PipelineConfig
from exoplanet_pipeline.batch import BatchRunConfig, run_raw_lightcurve_batch
from exoplanet_pipeline.synthetic import make_synthetic_transit_lc, make_synthetic_eb_lc, make_synthetic_blend_lc
from exoplanet_pipeline.final_outputs import generate_submission_package_outputs
```

```python
raws = [
    make_synthetic_transit_lc(tic_id=910001, period_days=3.0, depth_ppm=1200, noise_ppm=300, random_seed=1),
    make_synthetic_eb_lc(tic_id=920001, period_days=4.0, primary_depth_ppm=18000, secondary_depth_ppm=5000, random_seed=2),
    make_synthetic_blend_lc(tic_id=930001, period_days=3.4, observed_depth_ppm=1000, centroid_shift_pix=0.03, random_seed=3),
]
cfg = PipelineConfig(n_periods=800, detection_method='bls', make_plots=False, detection_use_variants=False)
bcfg = BatchRunConfig(output_dir=ROOT / 'outputs_parts_9_10_notebook', cache_dir=ROOT / 'outputs_parts_9_10_notebook/cache', resume=False)
result = run_raw_lightcurve_batch(raws, pipeline_config=cfg, batch_config=bcfg)
result['final_candidate_catalog'].head()
```

```python
paths = generate_submission_package_outputs(result['final_candidate_catalog'], ROOT / 'outputs_parts_9_10_notebook/submission_assets')
paths
```

The main deliverables are the final candidate catalog, target summary, failure log, candidate review markdown, visual summaries, and three-page report draft.
