# Parts 7–8 Demo: Uncertainty and Validation

This notebook shows how to run the synthetic injection-recovery validation and inspect the generated reports. Run it from the project root with `PYTHONPATH=src` or install the package in editable mode.

```python
from pathlib import Path
import pandas as pd
from exoplanet_pipeline.config import PipelineConfig
from exoplanet_pipeline.injection_recovery import default_injection_grid, run_injection_recovery_grid, summarize_injection_recovery
from exoplanet_pipeline.validation import validate_candidate_catalog, validation_report_to_markdown
```

```python
cfg = PipelineConfig(n_periods=300, n_durations=5, min_clean_points=300, detection_use_variants=False, make_plots=False)
specs = default_injection_grid(random_seed=42)[:6]
catalog = run_injection_recovery_grid(specs=specs, config=cfg, n_bootstrap=30)
catalog.head()
```

```python
summary = summarize_injection_recovery(catalog)
summary
```

```python
report = validate_candidate_catalog(catalog, label_col='label', pred_col='final_predicted_class')
print(validation_report_to_markdown(report))
```
